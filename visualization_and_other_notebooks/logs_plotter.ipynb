```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
num_epochs = 30
```

```python
fig = plt.figure(figsize=(10,5))
log_folders = ['logs_basic_vqa','logs']
legend_postfix = ['_basic-vqa','_attention-vqa']
colors_train = ['violet','darkred']
colors_val = ['orange','darkgreen']

plt.figure(figsize=(14,5))
for logi, logf in enumerate(log_folders):
    for phase in ['train', 'valid']:

        epoch = []
        loss = []
        acc = []

        for i in range(num_epochs):

            with open('./'+logf+'/{}-log-epoch-{:02d}.txt'.format(phase, i+1), 'r') as f:
                df = csv.reader(f, delimiter='\t')
                data = list(df)

            epoch.append(float(data[0][0]))
            loss.append(float(data[0][1]))
            acc.append(float(data[0][3]))

        plt.subplot(1, 2, 1)
        if phase == 'train':
            plt.plot(epoch, loss, label = phase+legend_postfix[logi], color = colors_train[logi], linewidth = 1.5)
        else:
            plt.plot(epoch, loss, label = phase+legend_postfix[logi], color = colors_val[logi], linewidth = 1.5)

        plt.xlabel('Epoch', fontsize = 20)
        plt.ylabel('Loss', fontsize = 20)
        plt.grid(axis='x',linestyle=':')
        
        plt.subplot(1, 2, 2)
        plt.tight_layout()

        if phase == 'train':
            plt.plot(epoch, acc, label = phase+legend_postfix[logi], color = colors_train[logi], linewidth = 1.5)
        else:
            plt.plot(epoch, acc, label = phase+legend_postfix[logi], color = colors_val[logi], linewidth = 1.5)

        plt.xlabel('Epoch', fontsize = 20)
        plt.ylabel('Accuracy', fontsize = 20)
        
        plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., prop={'size': 20})
        plt.grid(axis='x',linestyle=':')
        print(logf, "Valid epoch, min-loss, min-loss-accuracy, max-accuracy:",epoch[np.argmin(loss)],min(loss),acc[np.argmin(loss)], max(acc))

plt.grid(axis='x',linestyle=':')
plt.savefig('./png/train.png', dpi = fig.dpi)
```
